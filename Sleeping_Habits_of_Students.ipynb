```python
# import libraries

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
file2 = '/content/drive/MyDrive/Colab Notebooks/Data /kids sleep.csv'

df=pd.read_csv(file2)
df
```

```python
# sliced column - grade

grade = df['grade']
grade_array = grade.to_numpy()
grade_array
```

```python
# sliced column - hours needed

# convert sliced columns into numpy array

hoursNeeded=df['hours_needed']
hoursNeeded_array = hoursNeeded.to_numpy()
hoursNeeded_array
```

```python
# sliced column - hours averaged

# convert sliced columns into numpy array

hoursAveraged= df['hours_averaged']
hoursAveraged_array = hoursAveraged.to_numpy()
hoursAveraged_array
```

```python
# plot the bar graph
plt.bar(grade_array, hoursNeeded)

# Adding labels and a title
plt.xlabel('Grade')
plt.ylabel('Hours of Sleep Needed')
plt.title('The Number of Sleeping Hours Needed by Various Grade Levels')

# Ticks
plt.xticks(rotation=90)

# Show the chart
plt.show();
```

The graph shows a downward trend of the number of sleeping hours needed by students. As students get older, their number of sleeping hours needed decreases. Kindergartens require the most sleep, whereas 12th grade students require the least amount of sleep.

```python
# plot the bar graph
plt.bar(grade_array, hoursAveraged)

# Adding labels and a title
plt.xlabel('Grade')
plt.ylabel('Hours of Average Sleep')
plt.title('The Average Sleeping Hours by Various Grade Levels')

# Ticks
plt.xticks(rotation=90)

# Show the chart
plt.show();
```

The graph shows a downward trend. As students get older, the average sleeping hours they attain decreases. Kindergartens attain the most number of sleeping hours, whereas 12th grade students attain the least amount.

```python
# plot the bar graph
df.plot(x= 'grade', y=['hours_needed', 'hours_averaged'], kind='bar')

# Adding labels and a title
plt.xlabel('Grade')
plt.ylabel('Number of Hours')
plt.title('Sleeping Hours by Various Grade Levels')

# Ticks
plt.xticks(rotation=90)

# legend
plt.legend (bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the chart
plt.show();
```

The graph shows a downward trend for both sleeping hours needed and averaged number of hours. The gap between the number of sleeping hours needed and the average sleeping hours also increase as the year level increases.
